```python
# Load submissions CSV generated by Flask application

import csv

evaluations = []
with open('app/evaluations.csv', 'rb') as f:
    reader = csv.reader(f)
    evaluations = list(reader)

print len(evaluations)
evaluations[:5]
```

```python
# Get the unique participants

participants = set([evaluations[x][0] for x in range(1, len(evaluations))])
print len(participants)
participants
```

```python
# Count the number of votes and the success rate

positive_count = 0
negative_count = 0
scores = evaluations[1:]
scores = [x[1:] for x in scores]
for submission in scores:
    for entry in submission:
        if entry == 'True':
            positive_count += 1
        else:
            negative_count += 1
            
total = positive_count + negative_count
success_rate = positive_count / float(total)
print total
print positive_count
print negative_count
print success_rate
```
